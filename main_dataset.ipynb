```python
df_amazon=pd.read_csv("Amazon unique.csv")
df_instasports=pd.read_csv("Instasports unique.csv")
df_prokick=pd.read_csv("Prokick unique.csv")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sentence_transformers import SentenceTransformer
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import joblib
import streamlit as st
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
def preprocess_title(title):
    title = title.lower()
    # Remove flavors and colors (you can extend this list)
    colors_flavors = ["red", "blue", "green", "yellow", "black", "white", "watermelon", "chocolate"]
    for word in colors_flavors:
        title = title.replace(word, "")
    title = ''.join(e for e in title if e.isalnum() or e.isspace())
    return title.strip()
```

```python
df_amazon['Cleaned_Title'] = df_amazon['title'].apply(preprocess_title)
df_instasports['Cleaned_Title'] = df_instasports['title'].apply(preprocess_title)
df_prokick['Cleaned_Title'] = df_prokick['title'].apply(preprocess_title)
```

```python
df_amazon['title_vector'] = df_amazon['Cleaned_Title'].apply(lambda x: model.encode(x))
df_instasports['title_vector'] = df_instasports['Cleaned_Title'].apply(lambda x: model.encode(x))
df_prokick['title_vector'] = df_prokick['Cleaned_Title'].apply(lambda x: model.encode(x))
```

```python
def find_similar_products(prokick_vector, comparison_df, top_n=1, threshold=0.5):
    similarities = util.pytorch_cos_sim(prokick_vector, comparison_df['title_vector'].tolist())[0]
    sorted_indices = similarities.argsort(descending=True)
    similar_indices = [i for i in sorted_indices if similarities[i] > threshold][:top_n]
    return comparison_df.iloc[similar_indices]
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
```

```python
merged_df['amazon_available'] = merged_df['Amazon Title'].notna().astype(int)
merged_df['instasports_available'] = merged_df['Instasports Title'].notna().astype(int)
merged_df['Prokick Cleaned Title'] = merged_df['Prokick Cleaned Title'].str.lower()
```

```python
X_train, X_test, y_train_amazon, y_test_amazon = train_test_split(merged_df['Prokick Cleaned Title'], merged_df['amazon_available'], test_size=0.2, random_state=42)
X_train, X_test, y_train_instasports, y_test_instasports = train_test_split(merged_df['Prokick Cleaned Title'], merged_df['instasports_available'], test_size=0.2, random_state=42)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
train_encodings_amazon = tokenizer(list(X_train), truncation=True, padding=True)
test_encodings_amazon = tokenizer(list(X_test), truncation=True, padding=True)
train_encodings_instasports = tokenizer(list(X_train), truncation=True, padding=True)
test_encodings_instasports = tokenizer(list(X_test), truncation=True, padding=True)
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

```python
train_dataset_amazon = CustomDataset(train_encodings_amazon, list(y_train_amazon))
test_dataset_amazon = CustomDataset(test_encodings_amazon, list(y_test_amazon))
```

```python
train_dataset_instasports = CustomDataset(train_encodings_instasports, list(y_train_instasports))
test_dataset_instasports = CustomDataset(test_encodings_instasports, list(y_test_instasports))
```

```python
model_amazon = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
model_instasports = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
import torch

training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=8,   # batch size for training
    per_device_eval_batch_size=8,    # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
)
```

```python
trainer_amazon = Trainer(
    model=model_amazon,                         # the instantiated hf transformers model to be trained
    args=training_args,                         # training arguments, defined above
    train_dataset=train_dataset_amazon,         # training dataset
    eval_dataset=test_dataset_amazon             # evaluation dataset
)
```

```python
trainer_instasports = Trainer(
    model=model_instasports,                         
    args=training_args,                         
    train_dataset=train_dataset_instasports,         
    eval_dataset=test_dataset_instasports             
)
```

```python
trainer_amazon.train()
trainer_instasports.train()
```

```python
eval_result_amazon = trainer_amazon.evaluate()
eval_result_instasports = trainer_instasports.evaluate()

print(f"Amazon availability model evaluation: {eval_result_amazon}")
print(f"Instasports availability model evaluation: {eval_result_instasports}")
```

```python
def predict_availability(prokick_title):
    # Preprocess the input
    cleaned_title = prokick_title.lower()
    
    # Tokenize the input
    inputs = tokenizer(cleaned_title, return_tensors='pt', truncation=True, padding=True)

    # Predict availability on Amazon
    outputs_amazon = model_amazon(**inputs)
    logits_amazon = outputs_amazon.logits
    prediction_amazon = torch.argmax(logits_amazon, dim=1).item()

    # Predict availability on Instasports
    outputs_instasports = model_instasports(**inputs)
    logits_instasports = outputs_instasports.logits
    prediction_instasports = torch.argmax(logits_instasports, dim=1).item()

    # Fetch the corresponding titles if available
    amazon_title = None
    instasports_title = None
    
    if prediction_amazon == 1:
        amazon_title_row = merged_df[merged_df['Prokick Cleaned Title'] == cleaned_title]
        if not amazon_title_row.empty:
            amazon_title = amazon_title_row['Amazon Title'].values[0]
    
    if prediction_instasports == 1:
        instasports_title_row = merged_df[merged_df['Prokick Cleaned Title'] == cleaned_title]
        if not instasports_title_row.empty:
            instasports_title = instasports_title_row['Instasports Title'].values[0]

    return {
        'amazon_available': bool(prediction_amazon),
        'amazon_title': amazon_title,
        'instasports_available': bool(prediction_instasports),
        'instasports_title': instasports_title
    }

# Ask user for input and predict availability
prokick_title_input = input("Enter a product name you found on Prokick: ")
result = predict_availability(prokick_title_input)
print(f"Availability on Amazon: {result['amazon_available']}")
if result['amazon_available']:
    print(f"Amazon Title: {result['amazon_title']}")
print(f"Availability on Instasports: {result['instasports_available']}")
if result['instasports_available']:
    print(f"Instasports Title: {result['instasports_title']}")
```
