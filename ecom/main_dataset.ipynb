```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("Instasports.csv")
df2=pd.read_csv("Prokick.csv", encoding='latin1')
df3=pd.read_csv("Amazon.csv", encoding='latin1')
```

```python
df1.head()
```

```python

for df in [df1, df2, df3]:
    df['brand'] = df['brand'].astype(str)
    df['formatted_name'] = df['formatted_name'].astype(str)

# Add source columns to each dataframe
df1['available_on_instaclubsports'] = True
df2['available_on_prokicksports'] = True
df3['available_on_amazon'] = True

# Try merging again
try:
    merged_df = df1.merge(df2, on=['brand', 'formatted_name'], how='outer', suffixes=('_insta', '_proki'))
    merged_df = merged_df.merge(df3, on=['brand', 'formatted_name'], how='outer', suffixes=('', '_amzn'))
except ValueError:
    # If merge still fails, use concat
    print("Merge failed. Using concat instead.")
    merged_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)

# Fill NaN values in availability columns
for col in ['available_on_instaclubsports', 'available_on_prokicksports', 'available_on_amazon']:
    if col not in merged_df.columns:
        merged_df[col] = False
    merged_df[col] = merged_df[col].fillna(False)

# Select relevant columns
columns_to_keep = ['brand', 'formatted_name', 'title', 'offer_price', 'mrp',
                   'available_on_instaclubsports', 'available_on_prokicksports', 'available_on_amazon']
final_df = merged_df[columns_to_keep]

# Clean up the dataframe
final_df = final_df.fillna('')

# Display the first few rows of the resulting dataframe
print(final_df.head())
print(final_df.shape)
```
