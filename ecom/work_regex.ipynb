```python
# Merge Amazon and Prokick
merged_df = pd.merge(amazon[['normalized_title', 'offer_price', 'mrp']], 
                     Prokick[['normalized_title', 'offer_price','mrp']], 
                     on='normalized_title', 
                     suffixes=('_amazon', '_prokick'), 
                     how='outer')

# Merge the result with Instasports
merged_df = pd.merge(merged_df, 
                     instasports[['normalized_title', 'offer_price','mrp']], 
                     on='normalized_title', 
                     how='outer')

# Rename the instasports offer_price column
merged_df.rename(columns={'offer_price': 'offer_price_instasports'}, inplace=True)
```

```python
import re

def split_title(title):
    # Define patterns for different attributes
    color_pattern = r'\b(black|white|blue|red|green|orange|gold|grey|pink|yellow|cranberry|silver|navy|multicolor|pear|charcoal|lime|copper|cherry|maroon|beige)\b'
    pack_size_pattern = r'\b(pack of \d+)\b'
    limited_edition_pattern = r'\b(limited edition)\b'
    speed_pattern = r'\b(speed \d+)\b'
    
    # Extract attributes
    color = re.findall(color_pattern, title, re.IGNORECASE)
    pack_size = re.findall(pack_size_pattern, title, re.IGNORECASE)
    limited_edition = re.findall(limited_edition_pattern, title, re.IGNORECASE)
    speed = re.findall(speed_pattern, title, re.IGNORECASE)
    
    # Remove extracted attributes from the title
    clean_title = re.sub(color_pattern, '', title, flags=re.IGNORECASE)
    clean_title = re.sub(pack_size_pattern, '', clean_title, flags=re.IGNORECASE)
    clean_title = re.sub(limited_edition_pattern, '', clean_title, flags=re.IGNORECASE)
    clean_title = re.sub(speed_pattern, '', clean_title, flags=re.IGNORECASE)
    
    # Remove any remaining parentheses and extra spaces
    clean_title = re.sub(r'\([^)]*\)', '', clean_title)
    clean_title = re.sub(r'\s+', ' ', clean_title).strip()
    
    return {
        'clean_title': clean_title,
        'color': ', '.join(color) if color else None,
        'pack_size': pack_size[0] if pack_size else None,
        'limited_edition': 'Yes' if limited_edition else 'No',
        'speed': speed[0] if speed else None
    }

# Apply the function to the DataFrame
result = merged_df['normalized_title'].apply(split_title)

# Create new columns from the result
merged_df['clean_title'] = result.apply(lambda x: x['clean_title'])
merged_df['color'] = result.apply(lambda x: x['color'])
merged_df['pack_size'] = result.apply(lambda x: x['pack_size'])
merged_df['limited_edition'] = result.apply(lambda x: x['limited_edition'])
merged_df['speed'] = result.apply(lambda x: x['speed'])

# Display the result
print(merged_df[['clean_title', 'color', 'pack_size', 'limited_edition', 'speed']])
```

```python
def search_product(filtered_df, product_name):
    # Convert product name to lowercase for case-insensitive search
    product_name = product_name.lower()
    
    # Search for the product in the clean_title column
    matches = filtered_df[filtered_df['clean_title'].str.lower().str.contains(product_name, na=False)]
    
    if matches.empty:
        print(f"No products found matching '{product_name}'.")
        return
    
    for index, row in matches.iterrows():
        print(f"\nProduct: {row['clean_title']}")
        print("Availability:")
        
        # Check Amazon
        if pd.notna(row['offer_price_amazon']):
            print(f"  Amazon: Available at ₹{row['offer_price_amazon']:.2f} (MRP: ₹{row['mrp_amazon']:.2f})")
        else:
            print("  Amazon: Not available")
        
        # Check Prokick
        if pd.notna(row['offer_price_prokick']):
            print(f"  Prokick: Available at ₹{row['offer_price_prokick']:.2f} (MRP: ₹{row['mrp_prokick']:.2f})")
        else:
            print("  Prokick: Not available")
        
        # Check Instasports
        if pd.notna(row['offer_price_instasports']):
            print(f"  Instasports: Available at ₹{row['offer_price_instasports']:.2f} (MRP: ₹{row['mrp']:.2f})")
        else:
            print("  Instasports: Not available")
        
        print("\nAdditional Information:")
        if pd.notna(row['color']):
            print(f"  Color: {row['color']}")
        if pd.notna(row['pack_size']):
            print(f"  Pack Size: {row['pack_size']}")
        if pd.notna(row['speed']):
            print(f"  Speed: {row['speed']}")
        print(f"  Limited Edition: {row['limited_edition']}")

# Main program loop
while True:
    search_term = input("\nEnter a product name to search (or 'quit' to exit): ")
    if search_term.lower() == 'quit':
        break
    search_product(filtered_df, search_term)
```

```python
from tabulate import tabulate
from colorama import Fore, Back, Style, init
import random

# Initialize colorama for colored output
init(autoreset=True)

def search_product(filtered_df, product_name):
    product_name = product_name.lower()
    matches = filtered_df[filtered_df['clean_title'].str.lower().str.contains(product_name, na=False)]
    
    if matches.empty:
        return None
    
    results = []
    for index, row in matches.iterrows():
        result = {
            'product': row['clean_title'],
            'amazon_price': row['offer_price_amazon'] if pd.notna(row['offer_price_amazon']) else None,
            'prokick_price': row['offer_price_prokick'] if pd.notna(row['offer_price_prokick']) else None,
            'instasports_price': row['offer_price_instasports'] if pd.notna(row['offer_price_instasports']) else None,
            'color': row['color'] if pd.notna(row['color']) else None,
            'pack_size': row['pack_size'] if pd.notna(row['pack_size']) else None,
            'speed': row['speed'] if pd.notna(row['speed']) else None,
            'limited_edition': row['limited_edition']
        }
        results.append(result)
    
    return results

def generate_test_cases(filtered_df, num_cases=10):
    test_cases = []
    
    # Randomly select products
    sample_products = filtered_df.sample(n=num_cases)
    
    for _, row in sample_products.iterrows():
        search_term = row['clean_title']
        
        expected_result = {
            'product': row['clean_title'],
            'amazon_price': row['offer_price_amazon'] if pd.notna(row['offer_price_amazon']) else None,
            'prokick_price': row['offer_price_prokick'] if pd.notna(row['offer_price_prokick']) else None,
            'instasports_price': row['offer_price_instasports'] if pd.notna(row['offer_price_instasports']) else None,
            'color': row['color'] if pd.notna(row['color']) else None,
            'pack_size': row['pack_size'] if pd.notna(row['pack_size']) else None,
            'speed': row['speed'] if pd.notna(row['speed']) else None,
            'limited_edition': row['limited_edition']
        }
        
        test_case = {
            'search_term': search_term,
            'expected_results': [expected_result]
        }
        
        test_cases.append(test_case)
    
    # Add a test case for a nonexistent product
    test_cases.append({
        'search_term': 'nonexistent product xyz123',
        'expected_results': None
    })
    
    return test_cases

def run_accuracy_test(filtered_df, test_cases):
    total_tests = len(test_cases)
    correct_tests = 0
    results = []

    for i, test_case in enumerate(test_cases, 1):
        search_term = test_case['search_term']
        expected_results = test_case['expected_results']
        
        # Run the search
        actual_results = search_product(filtered_df, search_term)
        
        # Check if actual results match expected results
        is_correct = False
        if actual_results is None and expected_results is None:
            is_correct = True
        elif actual_results is not None and expected_results is not None:
            if len(actual_results) == len(expected_results):
                all_match = all(
                    actual['product'] == expected['product'] and
                    (abs(actual['amazon_price'] - expected['amazon_price']) < 0.01 if actual['amazon_price'] is not None and expected['amazon_price'] is not None else actual['amazon_price'] == expected['amazon_price']) and
                    (abs(actual['prokick_price'] - expected['prokick_price']) < 0.01 if actual['prokick_price'] is not None and expected['prokick_price'] is not None else actual['prokick_price'] == expected['prokick_price']) and
                    (abs(actual['instasports_price'] - expected['instasports_price']) < 0.01 if actual['instasports_price'] is not None and expected['instasports_price'] is not None else actual['instasports_price'] == expected['instasports_price']) and
                    actual['color'] == expected['color'] and
                    actual['pack_size'] == expected['pack_size'] and
                    actual['speed'] == expected['speed'] and
                    actual['limited_edition'] == expected['limited_edition']
                    for actual, expected in zip(actual_results, expected_results)
                )
                if all_match:
                    is_correct = True
        
        if is_correct:
            correct_tests += 1
            result = "Pass"
            color = Fore.GREEN
        else:
            result = "Fail"
            color = Fore.RED
        
        results.append([i, search_term[:30] + '...' if len(search_term) > 30 else search_term, result])
        
        print(f"\n{color}Test Case {i}: {search_term}")
        print(f"{color}Result: {result}")
        print(f"{Fore.YELLOW}Expected: {expected_results}")
        print(f"{Fore.CYAN}Actual: {actual_results}")

    accuracy = (correct_tests / total_tests) * 100
    return accuracy, results

# Main execution
if __name__ == "__main__":
    filtered_df = pd.read_csv('attribute.csv')  
    test_cases = generate_test_cases(filtered_df, num_cases=10)

    # Run the accuracy test
    accuracy, results = run_accuracy_test(filtered_df, test_cases)

    # Print summary table
    print("\n" + "="*50)
    print(f"{Fore.YELLOW}Test Results Summary:")
    print("="*50)
    print(tabulate(results, headers=['Test #', 'Search Term', 'Result'], tablefmt='pretty'))
    print("="*50)
    print(f"{Fore.GREEN}Model Accuracy: {accuracy:.2f}%")
    print("="*50)
```
